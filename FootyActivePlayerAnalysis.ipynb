# ACTIVE PLAYERS PER CLUB

```python
club_active_plyrs=fs['club'].value_counts().to_frame()
club_active_plyrs
```

```python
club_active_plyrs.plot(kind='bar',figsize=(16,5),xlabel='CLUBS',ylabel='NUMBER OF ACTIVE PLAYERS',title='ACTIVE PLAYERS V/S CLUB')
```

```python
print(f'Club with most active Player:{club_active_plyrs.idxmax()}')
print(f'Club with least active Player:{club_active_plyrs.idxmin()}')
```

```python
act_ply_age=fs['age'].value_counts().to_frame()
act_ply_age
```

```python
act_ply_age.plot(kind='bar',figsize=(16,5),xlabel='AGE',ylabel='NUMBER OF ACTIVE PLAYERS',title='ACTIVE PLAYERS V/S AGE')
```

```python
print(f'Age with most active Player:{act_ply_age.idxmax()}')
print(f'Age with least active Player:{act_ply_age.idxmin()}')
```

# ACTIVE PLAYERS PER COUNTRY


```python
fs
```

```python
fs['nationality']=fs['nationality'].str[-6:-4]
fs['nationality']
```

```python
nat_wise_act_plyrs=fs.groupby('nationality')['name'].agg('count').to_frame()
nat_wise_act_plyrs
```

```python
nat_wise_act_plyrs.plot(kind='bar',figsize=(16,5),xlabel='COUNTRY',ylabel='NUMBER OF ACTIVE PLAYERS',title='ACTIVE PLAYERS PER COUNTRY')
```

```python
print(f'Country with most active Player:{nat_wise_act_plyrs.idxmax()} ==> {nat_wise_act_plyrs.max()}')
print(f'Country with least active Player:{nat_wise_act_plyrs.idxmin()} ==> {nat_wise_act_plyrs.min()}')
```
